## Build a simple question-answering chatbot using a pre-trained language model.

```python
from transformers import pipeline

# Load the pre-trained Question-Answering pipeline
qa_model = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

# Provide a context (the document the bot will read)
context = """
The Solar System consists of the Sun and the objects that orbit it.
Among the eight planets, Jupiter is the largest, followed by Saturn.
Earth is the third planet from the Sun and the only known planet to support life.
"""
```

```python
def ask_bot(question):
    result = qa_model(question=question, context=context)
    return result['answer']

# Example usage
print(f"Answer: {ask_bot('Which planet is the largest?')}")
print(f"Answer: {ask_bot('Is Earth near the Sun?')}")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# 1. Load the model and tokenizer specifically for Seq2Seq
model_name = "facebook/blenderbot-400M-distill"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

def chat():
    print("Chatbot: Connection established. (Type 'exit' to quit)")

    while True:
        user_input = input("You: ")
        if user_input.lower() == 'exit':
            break

        # 2. Tokenize the input
        # We use return_tensors="pt" for PyTorch
        inputs = tokenizer(user_input, return_tensors="pt")

        # 3. Generate the response
        # We explicitly set max_new_tokens here to avoid the conflict you saw
        outputs = model.generate(**inputs, max_new_tokens=50)

        # 4. Decode and print
        reply = tokenizer.decode(outputs[0], skip_special_tokens=True)
        print(f"Chatbot: {reply}")

if __name__ == "__main__":
    chat()
```
